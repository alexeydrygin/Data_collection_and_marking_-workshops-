```python
import requests
import json
import pandas as pd
import datetime as dt
```

```python
url = 'https://api.binance.us/api/v3/klines'
symbol = 'BTCBUSD'
interval = '1h'
start = str(int(dt.datetime(2021,1,1).timestamp()*1000))
end = str(int(dt.datetime(2021,12,1).timestamp()*1000))
```

```python
params = {'symbol': symbol, 
          'interval': interval, 
          'startTime': start, 
          'endTime': end}
```

```python
response_text = requests.get(url, params= params).text
```

```python
# Отправка запроса API с помощью метода requests.get
response = requests.get(url, params=params)

# Проверка кода состояния ответа для подтверждения успешности запроса API
if response.status_code == 200:
    print("Успешный запрос API!")
else:
    print("Запрос API отклонен с кодом состояния:", response.status_code)

print(response_text)
```

```python
json_res = json.loads(response)
```

```python
json_res
```

```python
df = pd.DataFrame(json_res)

df.columns = ['datetime', 'open', 'high', 'low', 'close', 'volume','close_time', 'qav', 'num_trades','taker_base_vol', 'taker_quote_vol', 'ignore']
```
